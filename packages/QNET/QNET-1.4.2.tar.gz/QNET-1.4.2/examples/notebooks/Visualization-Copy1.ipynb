## draw paths

```python
def _line_patch(x1,y1, x2, y2):
    xm = (x1+x2)/2
    verts = [
        (x1, y1),
        (xm, y1),
        (xm, y2),
        (x2, y2),
    ]
    codes = [Path.MOVETO,
         Path.CURVE4,
         Path.CURVE4,
         Path.CURVE4,
         ]

    path = Path(verts, codes)

    patch = patches.PathPatch(path, facecolor='none', lw=2)
    return patch


    

plt.gca().add_patch(_line_patch(0,0,1,1))
plt.autoscale()
plt.show()
```
